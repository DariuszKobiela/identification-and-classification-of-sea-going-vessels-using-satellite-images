```python
import os

#C:\Users\user\Documents\identification-and-classification-of-sea-going-vessels-using-satellite-images\notebooks\TheCodingBug\YOLOv7_SHIPS\yolov7-custom2\data\val
ground_path = 'TheCodingBug/YOLOv7_SHIPS/yolov7-custom2/data/'
pictures_path = ground_path + 'train/images/'
labels_path = ground_path + 'train/labels/'

print(pictures_path)
print(os.path.exists(pictures_path))
pictures_list = os.listdir(pictures_path)
print(pictures_list)
print()

annotations_list = []
for picture in pictures_list:
    annot = picture.strip('.jpg') + '.txt'
    annotations_list.append(annot)
    
print(annotations_list)    
print()

labels_list = os.listdir(labels_path)
print(labels_list)
print()

for annotation in labels_list:
    if annotation not in annotations_list:
        print(annotation)
        os.remove(labels_path + annotation)
```
